```python
%%sql
INSERT INTO employees(id, name, department, position, salary, hire_date)
VALUES
(101,'Dolma', 'Engineering', 'Lead Engineer', 150000.0, '2024-09-25'),
(102,'Kaia', 'Engineering', 'Cloud Engineer', 135252.0, '2023-12-20');
```

```python
%%sql
UPDATE employees
SET name = 'Dolma Gurung'
WHERE id = 101;
```
