```python
%%sql
SELECT DISTINCT COUNT(*) AS number_of_people
FROM people;
```

```python
%%sql
SELECT sex,ROUND(AVG(age),2) AS average_age
FROM people
GROUP BY sex;
```

```python
%%sql
SELECT * FROM homes;
```

```python
%%sql
SELECT area, MAX(value) AS max_value
FROM homes
GROUP BY AREA
ORDER BY value DESC;
```

```python
%%sql
SELECT * FROM people
JOIN pets
ON people.name = pets.owner_name;
```

```python
%%sql
SELECT people.name, 
       people.sex,
       people.age, 
       pets.name AS pets_name,
       pets.type
FROM people
JOIN pets
ON people.name = pets.owner_name;
```

```python
%%sql
SELECT * FROM people
LEFT JOIN pets
ON people.name = pets.owner_name
UNION
SELECT * FROM people
RIGHT JOIN pets
ON people.name = pets.owner_name;
```
